```python
#exercise 1

patients = [
    {
        "name": "Alice",
        "age": 30,
        "weight": 60,
        "dosage":[
            {"time": "08:00", "dose": 10},
            {"time": "12:00", "dose": 40}
        ]
    },
    {
        "name": "Bob",
        "age": 16,
        "weight": 40,
        "dosage":[
            {"time": "08:00", "dose": 5},
            {"time": "12:00", "dose": -5}
        ]
    }
]

for patient in patients:
    name = patient["name"]
    age = patient["age"]
    weight = patient["weight"]

    if age >= 18:
        min_dose = 0.1 * weight
        max_dose = 0.5 * weight
    else:
        min_dose = 0.05 * weight
        max_dose = 0.3 * weight

    print(f"{name} (age: {age}, weight: {weight}kg, safe range: {min_dose:}_{max_dose:}mg)")

    valid = 0

    for record in patient["dosage"]:
        time = record["time"]
        dose = record["dose"]

        if dose <= 0:
            print("Invalid dose")
            continue
        valid += 1

        if min_dose <= dose <= max_dose:
            status = "safe"
        else:
            status = "unsafe"
        print(f"{time}:Dose={dose}mg _ {status}")
    print(f" Total valid dose: {valid}.")
```

```python
#exercise 3

patients = [
    {
        "name": "Alice",
        "age": 30,
        "weight": 60,
        "dosage":[
            {"time": "08:00", "dose": 10},
            {"time": "12:00", "dose": 40},
            {"time": "16:00", "dose": 35}
        ]
    },
    {
        "name": "Bob",
        "age": 16,
        "weight": 40,
        "dosage":[
            {"time": "08:00", "dose": 5},
            {"time": "12:00", "dose": 15},
            {"time": "16:00", "dose":"invalid"}
        ]
    }
]

for patient in patients:
    try:
        name = patient["name"]
        age = patient["age"]
        weight = patient["weight"]
        dosages = patient["dosage"]
        if not isinstance(age,int) or not isinstance(weight,(int, float)):
            raise ValueErroe("Invalid age or weight type.")

        if age >= 18:
            min_dose = 0.1 * weight
            max_dose = 0.5 * weight
        else:
            min_dose = 0.05 * weight
            max_dose = 0.3 * weight

        print(f"{name} (age: {age}, weight: {weight}kg, safe range: {min_dose:}_{max_dose:}mg)")

        valid_doses =[]
        safe_count = 0
        unsafe_count = 0

        for record in patient["dosage"]:
           try:
               time = record["time"]
               dose = record["dose"]

               if not isinstance(dose, int):
                   raise TypeError("invalid dose value.") 

               if dose <= 0:
                   print("Invalid dose")
                   continue
               valid_doses.append(dose)

               if min_dose <= dose <= max_dose:
                   print(f"{time}: dose={dose}mg _ safe")
                   safe_count += 1
               else:
                   print(f"{time}: dose={dose}mg _ unsafe")
                   unsafe_count += 1 
           except KeyError:
               print(" Error: missing 'time' or 'dose' Key in dosage record.")
           except TypeError:
               print("Error at 'time'.")
        
        avg_dose = sum(valid_doses) / len(valid_doses) if valid_doses else 0
        high_risk = (unsafe_count >= 2) or (avg_dose > max_dose)

        risk = "High risk" if high_risk else "low"
        reason = [{f"({unsafe_count} unsafe doses)" if unsafe_doses >= 2 else 0},
                 {f"(avg dose {avg_dose} > {max_dose})" if avg_dose > max_dose else None}]
        print(f"Risk Status:{risk}{reason}")
        total_valid = safe_count + unsafe_count
        if total_valid > 0:
            safe_p = 100 * safe_count / total_valid
            unsafe_p = 100 * unsafe_count / total_valid
            print(f" Distribution: safe: {safe_count} ({safe_p}%), unsafe: {unsafe_count} ({unsafe_p})")
        else:
            print(f" No valid doses")
    except KeyError as e:
        print(f"Error: missing {e}")
    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"Error: {e}")
```
